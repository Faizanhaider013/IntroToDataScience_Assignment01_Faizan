## Q1. Arithmetic and Assignment Operators


```python
# Q1: arithmetic and assignment
a = 12
b = 5

print("a =", a, "b =", b)
# arithmetic
print("addition:", a + b)
print("subtraction:", a - b)
print("multiplication:", a * b)
print("division:", a / b)         # float division
print("floor division:", a // b)  # integer division
print("modulus:", a % b)
print("exponent:", a ** b)

# assignment operators on a
a += 3
print("after a += 3 ->", a)
a -= 2
print("after a -= 2 ->", a)
a *= 2
print("after a *= 2 ->", a)
a /= 4
print("after a /= 4 ->", a)
```

## Q2. Comparison and Logical Operators

```python
# Q2: comparison and logical operators

x = int(input("Enter first integer: "))
y = int(input("Enter second integer: "))

print("x == y ->", x == y)
print("x != y ->", x != y)
print("x > y ->", x > y)
print("x < y ->", x < y)
print("x >= y ->", x >= y)
print("x <= y ->", x <= y)

# logical operators
print("x > 0 and y > 0 ->", x > 0 and y > 0)
print("x > 0 or y > 0 ->", x > 0 or y > 0)
print("not (x > y) ->", not (x > y))
```

## Q3. Bitwise Operators

```python
# Q3: bitwise operators
m = 6    # binary 110
n = 3    # binary 011

print("m & n (AND) ->", m & n)
print("m | n (OR)  ->", m | n)
print("m ^ n (XOR) ->", m ^ n)
print("~m (NOT)    ->", ~m)
print("m << 1 (left shift) ->", m << 1)
print("m >> 1 (right shift)->", m >> 1)
```

## Q4. Identity and Membership Operators

```python
# Q4: identity and membership
nums = [1, 2, 3, 4, 5]

a = nums
b = nums.copy()   # different object with same contents
c = a

print("a is b ->", a is b)     #  False
print("a is c ->", a is c)     # True

# membership
print("3 in nums ->", 3 in nums)
print("10 not in nums ->", 10 not in nums)
```

## Q5. String Operations

```python
# Q5: string operations
text = "Data Science is fun and powerful."

# length
print("length:", len(text))

# upper and lower
print("upper:", text.upper())
print("lower:", text.lower())

# find using 'in'
print("'fun' in text ->", 'fun' in text)

# replace
new_text = text.replace("fun", "awesome")
print("replaced:", new_text)

# slice (characters 0 to 9)
print("slice (0:10):", text[0:10])

# reverse using slicing
print("reverse:", text[::-1])
```
